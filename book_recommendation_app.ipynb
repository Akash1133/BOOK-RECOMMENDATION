```python
x=df1['User-ID'].value_counts()>200
y=x[x].index
```

```python
print(y)
df1=df1[df1['User-ID'].isin(y)]
```

```python
df1_with_df = df1.merge(df, on='ISBN')
df1_with_df.head()
```

```python
number_rating = df1_with_df.groupby('Book-Title')['Book-Rating'].count().reset_index()
number_rating.rename(columns= {'Book-Rating':'number_of_ratings'}, inplace=True)
final_rating = df1_with_df.merge(number_rating, on='Book-Title')
final_rating.shape
final_rating = final_rating[final_rating['number_of_ratings'] >= 15]
```

```python
final_rating.drop_duplicates(['User-ID','Book-Title'], inplace=True)
```

```python
book_pivot = final_rating.pivot_table(columns='Book-Title', index='User-ID', values="Book-Rating")
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
df=book_pivot
```

```python
df=df.corr(method='pearson')
```

```python
df
```

```python
def recommendation(book_name):
    x=df[book_name]
    x=x.sort_values(ascending=False)
    x=pd.DataFrame(x)
    return x.head(15)
```

```python
r=recommendation("Zombies of the Gene Pool")
```
